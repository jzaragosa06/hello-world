```python
import pandas as pd

# Sample DataFrame
data = {
    'A': [1, 2, 3, 4],
    'B': [10, 20, 30, 40],
    'C': [100, 200, 300, 400]
}

df = pd.DataFrame(data, index=['a', 'b', 'c', 'd'])
display(df)
```

# Loc - Label-Based Indexing

```python
row_b = df.loc['b']

#practically, the output is a pandas.Series object. Each row in df is represented as Sereis
row_b
```

```python
#selectding multiple rows
row_bc = df.loc[['a', 'b']]
row_bc
```

```python
#selecting rows and columns by label
# i.e., selecting specific cell using loc
entry = df.loc['b', 'B']
entry
```

```python
#slicing rows by label
#i.e., selecting rows by slicing it
slice = df.loc['b':'d']
#this returns a dataframe
slice

slice = df.loc['a':'c']
slice
```

```python
#conditional selection
condition = df.loc[df['B'] > 20]
condition
```

# iloc - Integer based indexing

```python
# Sample DataFrame
data = {
    'A': [1, 2, 3, 4],
    'B': [10, 20, 30, 40],
    'C': [100, 200, 300, 400]
}

df = pd.DataFrame(data)
display(df)
```

```python
row = df.iloc[1]
row
#the output is pandas.series object
```

```python
#selecting multiple rows
rows = df.iloc[[1, 2]]
rows
```

```python
#slicing
#it is includsive
rows = df.iloc[0:2 + 1]
rows
```

```python
#selecting specific value
value = df.iloc[1, 2]
value
```
